```python
import pandas as pd
import numpy as np
```

```python
no_fraud_data = pd.read_csv('./source/no_fraud_data.csv')
no_fraud_data = no_fraud_data.rename(columns={'value':'amount'})
no_fraud_data
```

```python
fraud_data = pd.read_csv('./source/fraud_data.csv')
fraud_data = fraud_data.rename(columns={'source':'from','target':'to'})
fraud_data
```

```python
no_fraud_data['amount'] = no_fraud_data['amount']/1e+18
fraud_data['isPhi'] = 1
no_fraud_data['isPhi'] = 0

data = pd.concat([no_fraud_data, fraud_data], axis=0)
data
```

```python
counts_from = data['from'].value_counts()
counts_from = counts_from.reset_index()
counts_from = counts_from.rename(columns={'index':'address','from':'send_occurances'})
counts_from
```

```python
counts_to = data['to'].value_counts()
counts_to = counts_to.reset_index()
counts_to = counts_to.rename(columns={'index':'address','to':'receive_occurances'})
counts_to
```

```python
merged_counts = pd.merge(counts_from, counts_to, on='address', how='outer')
merged_counts
```

```python
merged_counts['send_occurances'].fillna(0, inplace=True)
merged_counts['receive_occurances'].fillna(0, inplace=True)
merged_counts
```

```python
merged_counts['involved_transactions'] = merged_counts['send_occurances']+merged_counts['receive_occurances']
merged_counts['ratio_occurances'] = merged_counts['receive_occurances']/merged_counts['send_occurances']
merged_counts
```

```python
amount_from = data.groupby('from')['amount'].sum().reset_index()
amount_from = amount_from.rename(columns={'from':'address','amount':'amount_send'})
amount_to = data.groupby('to')['amount'].sum().reset_index()
amount_to = amount_to.rename(columns={'to':'address','amount':'amount_receive'})

merged_amount = pd.merge(amount_from, amount_to, on='address', how='outer')
merged_amount
```

```python
merged_amount['amount_send'].fillna(0, inplace=True)
merged_amount['amount_receive'].fillna(0, inplace=True)

merged_amount['amount_ratio'] = merged_amount['amount_receive']/merged_amount['amount_send']
merged_amount['amount_ratio'].fillna(0, inplace=True)

merged_amount['balance'] = merged_amount['amount_receive']-merged_amount['amount_send']
merged_amount
```

```python
merged_amount_and_counts = pd.merge(
	merged_amount,merged_counts,
	on='address',
	how='outer'
	)
merged_amount_and_counts
```

```python
merged_amount_and_counts['means_send'] = merged_amount_and_counts['amount_send']/merged_amount_and_counts['send_occurances']
merged_amount_and_counts['means_receive'] = merged_amount_and_counts['amount_receive']/merged_amount_and_counts['receive_occurances']
merged_amount_and_counts.fillna(0, inplace=True)

merged_amount_and_counts['means_ratio'] = merged_amount_and_counts['means_receive']/merged_amount_and_counts['means_send']
merged_amount_and_counts.fillna(0, inplace=True)
merged_amount_and_counts
```

```python
fraud_from = data[['from','isPhi']]
fraud_to = data[['to','isPhi']]

fraud_from = fraud_from.rename(columns={'from':'address'})
fraud_to = fraud_to.rename(columns={'to':'address'})
fraud_merged =  pd.concat([fraud_from, fraud_to], axis=0)

fraud_merged = fraud_merged.drop_duplicates()
fraud_merged
```

```python
data = pd.merge(
	merged_amount_and_counts,
	fraud_merged,
	on='address',
	how='outer'
)
data
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
x= keras.Input(shape=(1,2))
x1 = keras.Input(shape=(1, 11))

print(x.shape)
print(x1.shape)

y = keras.layers.LSTM(128, return_sequences=True)(x)
y = keras.layers.LSTM(128, return_sequences=True)(y)


y1 = keras.layers.Dense(128, activation='relu')(x1)
y1 = keras.layers.Dense(256, activation='relu')(y1)
y1 = keras.layers.Dense(128, activation='relu')(y1)

join2 = keras.layers.concatenate([y,y1], axis=1)

y2 = keras.layers.Dense(32)(join2)
y2 = keras.layers.Dropout(0.5)(y2)
y2 = keras.layers.Dense(32)(y2)
y2 = keras.layers.Reshape((-1,))(y2)
final1 = keras.layers.Dense(1, activation='sigmoid')(y2)
final2 = keras.layers.Dense(1, activation='sigmoid')(y2)

model = tf.keras.Model(inputs=[x,x1], outputs=[final1, final2])
```

```python
[x.shape for x in model.inputs]
```

```python
[x.shape for x in model.outputs]
```

```python
model.compile(optimizer='adam',loss = 'binary_crossentropy',metrics=['acc','mse'])
```

```python
model.fit([w1_input_train, w2_input_train], [w1_output_train, w2_output_train], batch_size=128, epochs=100)
```

```python
check = model.evaluate([w1_input_test, w2_input_test], [w1_output_test, w2_output_test])
print (check)
print('Test loss:', check[0])
print('Test accuracy:', check[1])
```
